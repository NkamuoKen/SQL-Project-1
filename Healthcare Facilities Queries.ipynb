```python
# Query 11 - Identify facilities that have more pharmacists than the average number of pharmacists in their respective states.

query_11 = """
WITH state_avg_pharmacists AS (
    SELECT state, AVG(pharmacists) AS avg_pharmacists
    FROM healthcare_facility_data
    GROUP BY state
)
SELECT f.facility_name, f.state, f.pharmacists
FROM healthcare_facility_data f
JOIN state_avg_pharmacists s
ON f.state = s.state
WHERE f.pharmacists > s.avg_pharmacists;
"""
```

```python
cur.execute(query_11)
```

```python
facilities_with_more_pharmacists_than_average = cur.fetchall()
```

```python
facilities_with_more_pharmacists_than_average
```

```python
# Query 12 - Calculate the median number of nurses for each facility level and order the results by the median in 
# descending order.

query_12 = """
SELECT facility_level, percentile_cont(0.5) WITHIN GROUP (ORDER BY nurses) AS median
FROM healthcare_facility_data
GROUP BY facility_level
ORDER BY median DESC;
"""
```

```python
cur.execute(query_12)
```

```python
median_number_of_nurses = cur.fetchall()
```

```python
median_number_of_nurses
```

```python
# Query 13 - Find the facility with the highest number of healthcare professionals in each state.

query_13 = """
WITH RankedHealthcareProfessionals AS (
    SELECT facility_name, state,
        doctors + pharmacists + dentist + pharmacy_technicians + nurses + lab_scientists + midwifes + 
        lab_technicians + nurse_midwife + him_officers + community_health_officer + 
        community_extension_workers + jun_community_extension_worker + dental_technicians + 
        env_health_officers AS total_healthcare_professionals,
        ROW_NUMBER() OVER (PARTITION BY state ORDER BY doctors + pharmacists + dentist + 
                           pharmacy_technicians + nurses + lab_scientists + midwifes + 
                           lab_technicians + nurse_midwife + him_officers + community_health_officer + 
                           community_extension_workers + jun_community_extension_worker + 
                           dental_technicians + env_health_officers DESC) AS rn
    FROM
        healthcare_facility_data
)
SELECT 
    facility_name,
    state,
    total_healthcare_professionals
FROM
    RankedHealthcareProfessionals
WHERE
    rn = 1;
"""
```

```python
cur.execute(query_13)
```

```python
facilities_with_more_professionals_per_state = cur.fetchall()
```

```python
facilities_with_more_professionals_per_state
```

```python
# Query 14 - Rank the facilities within each state based on the density of healthcare professionals 
# (number of professionals per bed), and select the top 3 facilities in each state.

query_14 = """
WITH RankedFacilities AS (
    SELECT
        facility_name,
        state,
        doctors + pharmacists + dentist + pharmacy_technicians + nurses + lab_scientists + midwifes + 
        lab_technicians + nurse_midwife + him_officers + community_health_officer + 
        community_extension_workers + jun_community_extension_worker + dental_technicians + 
        env_health_officers AS total_healthcare_professionals,
        beds,
        ROW_NUMBER() OVER (PARTITION BY state ORDER BY 
            CASE WHEN beds <> 0 THEN (doctors + pharmacists + dentist + pharmacy_technicians + nurses + 
        lab_scientists + midwifes + lab_technicians + nurse_midwife + him_officers + community_health_officer + 
        community_extension_workers + jun_community_extension_worker + dental_technicians + 
        env_health_officers) / beds END DESC) AS rank_within_state
    FROM
        healthcare_facility_data
)
SELECT
    facility_name,
    state,
    total_healthcare_professionals
FROM
    RankedFacilities
WHERE
    rank_within_state <= 3
    AND total_healthcare_professionals > 0;
"""
```

```python
cur.execute(query_14)
```

```python
rank_of_facilities_per_state = cur.fetchall()
```

```python
rank_of_facilities_per_state
```
